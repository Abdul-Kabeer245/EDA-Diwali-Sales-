This is a EDA practice project of diwali sales analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = 'unicode_escape')
df.head()
```

```python
df.info()
```

```python
df.drop(columns =['Status', 'unnamed1'], axis =1 , inplace = True)
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
df.shape
```

```python
df.describe()
```

<h1>EDA</h1>


<h3>Gender</h3>

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

From above graphs we can see that most of the buyer are females and even the purchasing power of female is greater than the men.

<h3>Age</h3>

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)['Amount'].sum().sort_values(by = "Amount", ascending = False)
```

```python
sns.barplot(x= 'Age Group', y = 'Amount',data = sales_age)
```

From above graph we can see that most of age group between 26-35 year female.

<h3>State</h3>

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(15, 5)})
sns.barplot(data = sales_state, x = 'State', y= 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)
sns.barplot(data = sales_state, x= 'State', y = 'Amount')
```

From above graph we can see that most orders and total sales/amount are from Uttar padesh, Maharashtra and Karnataka respectively.

<h3>Marital Status</h3>

```python
sns.set(rc = {'figure.figsize': (6,5)})
ax = sns.countplot(data = df, x = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(data = sales_marital_status, x = 'Marital_Status', y = 'Amount',hue = 'Gender')
```

From the above graph we see that most of the buyers are married women and they also have high purchasing power.

<h3>Occupation</h3>

```python
sns.set(rc = {'figure.figsize': (20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(data = sales_occupation, x= 'Occupation', y = 'Amount')
```

From the above graph we can see that most of the buyers work in IT sector, Healthcare, Aviation 

<h3>Product Category</h3>

```python
sns.set(rc ={'figure.figsize': (24,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

From the above graph we can see that most sold product category was food, clothing and footware.

<h1>Conclusion</h1>

Married women of age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare, and Aviation are more likely to buy products from Food, Clothing and Electroincs category.
